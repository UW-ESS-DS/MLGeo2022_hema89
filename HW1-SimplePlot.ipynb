### Importing modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Initiating variables

x = np.arange(1, 102, 2)
y = x * 2
```

```python
# set default parameters
params = {'legend.fontsize': 14, \
          'xtick.labelsize':14, \
          'ytick.labelsize':14, \
          'font.size':14}
plt.rcParams.update(params)
```

```python
#Plotting the variables
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))  # 1 row, 2 co
ax1.plot(x, y,'r.')
ax2.plot(y, -x, '.')
ax2.set_xlabel('Y')
ax1.set_xlabel('X')
ax2.set_ylabel('-X')
ax1.set_ylabel('Y')
ax1.set_title('Y vs X')
ax2.set_title('-X vs Y')
ax1.grid(True)
ax2.grid(True)
plt.savefig('plot_test.png')
```
